```python
multpl_stock_daily_returns = multpl_crypto['Adj Close'].pct_change()
multpl_stock_monthly_returns = multpl_crypto['Adj Close'].resample('M').ffill().pct_change()
```

```python
fig = plt.figure()
(multpl_stock_monthly_returns + 1).cumprod().plot()
plt.show()
```

```python
mean_df = pd.DataFrame(multpl_stock_monthly_returns.mean())
mean_df
```

```python
mean_df.plot.barh()
plt.title('Average Return')
plt.xlabel('Return/month (x100)')
plt.legend('')
```

```python
Std_df = pd.DataFrame(multpl_stock_monthly_returns.std())
Std_df
```

```python
Std_df.plot.barh()
plt.title('Volatility')
plt.xlabel('Standard Deviation')
plt.legend('')
```

```python
corr_df = (multpl_stock_monthly_returns.corr())
corr_df
```

```python
corrMatrix = corr_df.corr()
sn.heatmap(corrMatrix, 
           annot=True,
           vmin=-1, vmax=1, center=0,
           cmap=sn.diverging_palette(20, 220, n=200),
           square=True
          )
plt.show()
```

```python
tickers = ["AVGO"]
AVGO_stock = web.get_data_yahoo(tickers,
start = "2020-09-01",
end = "2021-01-17")
```

```python
multpl_stock_daily_returns2 = AVGO_stock['Adj Close']
multpl_stock_monthly_returns2 = AVGO_stock['Adj Close'].resample('M').ffill()

clean_AVGO = multpl_stock_monthly_returns2.dropna()
print(clean_AVGO)
#multpl_stock_monthly_returns2
```

```python

x = [i for i in np.arange(1,6,1)]
y = clean_AVGO['AVGO']

fig = plt.scatter(x, y)
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.show()
```

```python
multpl_stock_daily_returns2 = AVGO_stock['Adj Close']
multpl_stock_monthly_returns2 = AVGO_stock['Adj Close'].resample('M').ffill()

AVGO_stock2 = multpl_stock_daily_returns2.dropna()
AVGO_stock2
```

```python
x = [i for i in np.arange(1,96,1)]
y = AVGO_stock2['AVGO']

fig = plt.scatter(x, y, c='darkblue', s=5)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.xlabel('Time period - Days')
plt.ylabel('Stock Price')
plt.title('AVGO')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as web
from scipy.stats import linregress
import pylab
import mplfinance as mpf
import seaborn as sn
```

```python
tickers = ["ADI", "AMT", "CCI", "AVGO", "QCOM", "TMUS"]
multpl_stocks = web.get_data_yahoo(tickers,
start = "2016-01-01",
end = "2021-01-08")
```

```python
multpl_stocks
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)
ax6 = fig.add_subplot(326)

ax1.plot(multpl_stocks['Adj Close']['ADI'])
ax1.set_title("ADI")
ax2.plot(multpl_stocks['Adj Close']['AMT'])
ax2.set_title("AMT")
ax3.plot(multpl_stocks['Adj Close']['CCI'])
ax3.set_title("CCI")
ax4.plot(multpl_stocks['Adj Close']['AVGO'])
ax4.set_title("AVGO")
ax5.plot(multpl_stocks['Adj Close']['QCOM'])
ax5.set_title("QCOM")
ax6.plot(multpl_stocks['Adj Close']['TMUS'])
ax6.set_title("TMUS")

plt.tight_layout()
plt.show()
```

```python
multpl_stock_daily_returns = multpl_stocks['Adj Close'].pct_change()
multpl_stock_monthly_returns = multpl_stocks['Adj Close'].resample('M').ffill().pct_change()
```

```python
fig = plt.figure()
(multpl_stock_monthly_returns + 1).cumprod().plot()
plt.show()
```

```python
mean_df = pd.DataFrame(multpl_stock_monthly_returns.mean())
mean_df
```

```python
mean_df.plot.barh()
plt.title('Average Return')
plt.xlabel('Return/month (x100)')
plt.legend('')
```

```python
Std_df = pd.DataFrame(multpl_stock_monthly_returns.std())
Std_df
```

```python
Std_df.plot.barh()
plt.title('Volatility')
plt.xlabel('Standard Deviation')
plt.legend('')
```

```python
corr_df = (multpl_stock_monthly_returns.corr())
corr_df
```

```python
corrMatrix = corr_df.corr()
sn.heatmap(corrMatrix, 
           annot=True,
           vmin=-1, vmax=1, center=0,
           cmap=sn.diverging_palette(20, 220, n=200),
           square=True
          )
plt.show()
```
